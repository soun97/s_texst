```python
종목코드별수익률 = [
    {"종목코드" : "A001", "날짜목록" : ["2023-01-01", ""], "수익률" : [3, 2, 2, 3]},
    {"종목코드" : "B001", "날짜목록" : ["2023-01-01", ""], "수익률" : [2, 2, 2, 3]},
    {"종목코드" : "C001", "날짜목록" : ["2023-01-01", ""], "수익률" : [6, 2, 2, 3]},
]



합계 = [0, 0, 0, 0]

for i in 종목코드별수익률:
    # i : {"종목코드" : "A001", "날짜목록" : ["2023-01-01", ""], "수익률" : [3, 2, 2, 3]}
    
    수익률 = i.get('수익률')
    # 수익률 :  [3, 2, 2, 3]
    
    for idx, value in enumerate(수익률):
        # idx : 0, value : 3
        # idx : 1, value : 2
        # ..
        합계[idx] = 합계[idx] + value
 

for x in 합계:
    평균 = x / len(종목코드별수익률)
    print(평균)
```

```python

tax = 0.99985
stock_tax = 0.9977

def change_rate(a,b):
    if a==0:
        return 0
    else:
        return str(round((b-a)*100/a,2)) + "%" #cal(a, b)에 return한 값이 들어간다.

def earning_rate(a,b):
    if a==0:
        return 0
    else:
        return str(round((b*tax*stock_tax-a*tax)*100/a*tax,2)) + "%" #cal(a, b)에 return한 값이 들어간다.

file = 'C:/Users/soun/Documents/soun_pycharm/append2023-09-13.csv'
df2 = pd.read_csv(file,encoding='cp949')
sigalist = df2['시가']
jongalist= df2['종가']
resultlist = []
resultlist_b = []

for i in range(len(sigalist)):
    a = sigalist[i]
    b = jongalist[i]
    result = change_rate(a,b)
    result_b = earning_rate(a,b)
    resultlist.append(result)
    resultlist_b.append(result_b)
# print(resultlist)

df2= pd.DataFrame({'변화율': resultlist, '수익률':resultlist_b})#dataframe 만드는 방법:
# print(df,df2)
pd.concat((df,df2), axis = 1).to_csv('C:/Users/soun/Documents/soun_pycharm/append2023-09-13.csv', encoding='cp949', index=False)#dataframe끼리 더하는 방법
rea
```
